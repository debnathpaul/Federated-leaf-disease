# Federated Learning Simulation - Complete Results DemoThis notebook demonstrates a complete federated learning simulation for leaf disease detection using a 3-tier hierarchical architecture.## Project Overview- **Task**: Leaf disease classification (6 classes)- **Model**: Lightweight CNN trained from scratch- **Architecture**: Edge Nodes (6 farms) -> Regional Aggregators (3 regions) -> Global Server- **Rounds**: 3 FL rounds with sequential farm training- **Data Distribution**: Non-IID split across farms to simulate real-world conditions

```python
import osimport numpy as npimport pandas as pdimport matplotlib.pyplot as pltfrom PIL import Image# Setupproject_root = os.path.abspath('..')results_dir = os.path.join(project_root, 'results')print(f'Project root: {project_root}')print(f'Results directory: {results_dir}')print(f'Files available: {os.listdir(results_dir)}')
```

## 1. Simulation Data Summary### Farm Configuration- **6 farms** distributed across **3 regions** (A, B, C)- Each region has 2 farms with a regional aggregator- Farms train locally on non-IID data splits### FL Rounds- **Round 1-3**: Each farm trains for 5 local epochs- Farms send weights to regional aggregators- Aggregators perform FedAvg and send to global server- Global server updates model and sends back to aggregators
